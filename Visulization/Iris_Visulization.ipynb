```python
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
irs= load_iris()
```

```python
dir(irs)
```

```python
irs.feature_names
```

```python
irs.target
```

```python
irs.target_names
```

```python
df = pd.DataFrame(irs.data,columns=irs.feature_names)
df.sample(7)
```

```python
df["Target"] = irs.target
```

```python
df
```

```python
flower_names = []
for i in irs.target:
    flower_names.append(irs.target_names[i])
#flower_names
```

```python
df["Flower_name"] = flower_names
df.sample(10)
```

```python
df.shape
```

```python
df.info()
```

```python
irs.target_names
```

```python
df_setosa=df[df.Flower_name == "setosa"]
df_versicolor=df[df.Flower_name =="versicolor"]
df_virginica=df[df.Flower_name =="virginica"]
```

```python
print(df_setosa.shape)
print(df_versicolor.shape)
print(df_virginica.shape)
```

# Univariate Analysis

```python
# lets yake petal_length
plt.figure(figsize=(15,10))
plt.plot(df_setosa['sepal length (cm)'],np.zeros_like(df_setosa['sepal length (cm)']),'r+', markersize=22,label="setosa")
plt.plot(df_virginica['sepal length (cm)'],np.zeros_like(df_virginica['sepal length (cm)']),'g+', markersize=16,label='virginica')
plt.plot(df_versicolor['sepal length (cm)'],np.zeros_like(df_versicolor['sepal length (cm)']),'b+', markersize=18,label='versicolor')

plt.xlabel("Flowers")

plt.legend()
```

# Bivariate Analysis

```python
sn.FacetGrid(df,hue='Flower_name',size=7).map(plt.scatter,"sepal length (cm)","sepal width (cm)").add_legend();
```
