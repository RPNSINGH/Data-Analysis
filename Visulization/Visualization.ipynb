# seaborn.jointplot
Draw a plot of two variables with bivariate and univariate graphs.

```python
sn.jointplot(data=df,x="tip", y="total_bill")
```

```python
sn.jointplot(data=df, x="total_bill", y="tip",kind='hex')
```

```python
sn.jointplot(data=df, x="tip", y="total_bill",kind='reg')
```

# Multivariate EDA

# seaborn.pairplot() 
To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. 

```python
sn.pairplot(df)
```

```python
sn.pairplot(df,hue='smoker')
```

# seaborn.distplot

It is used basically for univariant set of observations and visualizes it through a histogram

```python
sn.distplot(df['tip'])
```

```python
sn.distplot(df['tip'],kde=False)
```
