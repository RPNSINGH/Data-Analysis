<h3 style="color:navy"> we can write 5th January ,2017 in too many ways : 

<h3 style="color:orange"> Now we will use 'pd.to_datetime()'  method to get  date with a proper format

```python
import pandas as pd
```

```python
dates=['2017-01-05','2017/01/05','01-05-2017','jan 5,2017','20170105']

pd.to_datetime(dates)
```

```python
# it can also handle time also
dates=['2017-01-05 4:30 pm','2017/01/05','01-05-2017 4:30:00','jan 5,2017','20170105']

pd.to_datetime(dates)
```

<h3 style="color:navy">    In US     : mm/dd/yyyy  </h3>
<h3 style="color:purple"> In Europe : dd/mm/yyyy

```python
pd.to_datetime('5/1/2017',dayfirst=True)
```

```python
dates=['sdsfd','2017-01-05','2017/01/05','01-05-2017','jan 5,2017','20170105','dfgf']
pd.to_datetime(dates)
# this will throw an error . 
```

```python
dates=['sdsfd','2017-01-05','2017/01/05','01-05-2017','jan 5,2017','20170105','dfgf']
pd.to_datetime(dates,errors='ignore')
#If 'ignore', then invalid parsing will return the input.
```

```python
#If 'coerce', then invalid parsing will be set as N
pd.to_datetime(dates,errors='coerce')
```
